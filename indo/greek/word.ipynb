Ναι, το νερό.

Μαρία, το καρότο.

Άννα και Μαρκ.

Ναι, το ροζ βάζο.

Το αβοκάντο είναι ροζ.

Είναι νερό.

Το ρύζι είναι ροζ.

Το ροζ νερό.

Είναι νερό και καρότο.

Ναι, το βίντεο.

Αυτός κάνει ρύζι.

Σνακ με ούζο.

Βίντεο και σνακ.

Μουσείο με σινεμά.

Είναι το μουσείο.

Το τοστ και το καρότο.

Σουσάνα, το τοστ.

Κουσκούς με καρότο.
