Một cô gái. 

Tôi muốn học. 

Người phụ nữ là bạn. 

Tôi là một người đàn ông. 

Tôi uống nước. 

Cô ấy là người. 

Người đàn ông uống nước ép tào. 

Tôi là một cô gái. 

Bạn là một bé gái. 

Bé gái muốn học. 

Cô gái ăn tào. 

Bé gái muốn uống nước. 

Cậu bé muốn học. 

Đứa trẻ học, anh ấy uống nước. 

Một cái găng tay.

An ở nhà ga.

Con dơi ở nhà ga.

Cái găng tay ở nhà ga.

Một cái ca dơ.

Một con dơi dơ.

Con gà ăn con cá.

Con gà cắn Ăn.

Con cá cắn.

An ăn đu đủ.

Con dê ăn đu đủ.

Dì có một cái ô.

Dì có một con dơi.

Con dê và con gà ăn.
