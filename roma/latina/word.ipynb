Quis custodiet ipsos custodes?

Quod erat demonstrandum.

Exempli gratia.

Puer sum.

Ego sum vir.

Ego sum Corinna.

Sum femina.

Marcus est vir.

Livia est femina.

Stephanus est puer.

Puella sum.

Vir dormit.

Quis dormit?

Quis sum ego?

Livia non est vir.

Femina non sum, sed vir.

Puer non dormit, sed puella dormit.

Tu es vir.

Quis es tu?

Ego non sum Stephanus.

Quis est is?

Quis est ea?

Ea est puella.

Ego domi sum.

Femina studet.

Puella scribit.

Quis est in urbe?

Femina studet et scribit.

Marcus scribit et non dormit.

Is domi dormit.

Is non est Marcus.

Non puer, sed vir scribit.

Mater scribit.

Tu es pater.

Ea est soror.

Soror in urbe est.

Frater studet.

Frater domi dormit.

Soror in urbe est, mater domi est.

Corinna domi non dormit.

Tu femina non es, sed puella es.

Tu non es pater.

Salve!

Salvete, Marce et Livia.

Salvete, Livia et Corinna.

Quid agit Livia?

Quid agit Stephanus?

Nomen ei est Livia.

Quid est ei nomen?

Quid tibi est nomen?

Quid est mihi nomen?

Nomen mihi est Marcus.

Salve, quid tibi est nomen?

Quid habet Marcus?

Livia se bene habet.

Livia se male habet.

Quomodo femina se habet?

Male me habeo.

Stephanus bene se habet.

Me male habeo.

Me bene habeo.

Quomodo me habeo?

Quomodo Corinna se habet?

Quomodo te habes?

Ubi Marcus dormit?

Ubi habitas?

Ubi est Italia?

Habitasne in urbe?

Ita, me bene habeo.

Roma est in Italia.

Minme! Marcus domi dormit.

Habitasne in urbe?

Romae habito.

Romae habitas.

Ita, ego Romae habito.

In Italia habitatis.

Ubi tu habitas?

Quomodo te habes Marce?

Livia, quomodo tu te habes?

Quis est is?

Stephanus studet et scribit.

Livia in urbe studet.

Livia non in urbe studet.

Soror in urbe est, mater domi est.
